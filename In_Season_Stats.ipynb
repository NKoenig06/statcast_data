## Hitter Stats

```python
hitter_stats_2024['batter_float'] = hitter_stats_2024['batter'].astype(float)
mapping_needed['MLBID'] = mapping_needed['MLBID'].astype(float)
```

```python
season_2024 = hitter_stats_2024.merge(mapping_needed,left_on='batter_float',right_on='MLBID')
```

```python
# crosstab
season_2024_events_df = pd.crosstab(season_2024['PLAYERNAME'],season_2024['events'])

season_2024_events_df['plate_appearances'] = season_2024_events_df[['walk','field_out','grounded_into_double_play', 'strikeout',
                                                                                'single','double','force_out','field_error','hit_by_pitch','home_run',
                                                                                'strikeout_double_play','triple', 'fielders_choice','sac_fly',
                                                                                'sac_bunt','double_play','fielders_choice_out']].sum(axis=1)
season_2024_events_df['at_bats'] = season_2024_events_df[['field_out','grounded_into_double_play','strikeout',
                                                                                'single','double','force_out','field_error','home_run',
                                                                                'strikeout_double_play','triple','fielders_choice','double_play','fielders_choice_out'
                                                                                ]].sum(axis=1)
```

```python
season_2024_events_df['walk_rate'] = (season_2024_events_df['walk']/season_2024_events_df['plate_appearances']).round(3)
season_2024_events_df['strikeout_rate'] = (season_2024_events_df['strikeout']/season_2024_events_df['plate_appearances']).round(3)
season_2024_events_df['OBP'] = (season_2024_events_df[['walk','single','double','hit_by_pitch','home_run','triple'
                                                                         ]].sum(axis=1)/season_2024_events_df['plate_appearances']).round(3)
season_2024_events_df['BA'] = (season_2024_events_df[['single','double','home_run','triple'
                                                                         ]].sum(axis=1)/season_2024_events_df['at_bats']).round(3)
```

```python
## Get Barrel %
# crosstab
season_2024_barreled_df = pd.crosstab(season_2024['PLAYERNAME'],season_2024['launch_speed_angle'])
season_2024_barreled_df['lsa_events'] = season_2024_barreled_df[[1,2,3,4,5,6]].sum(axis=1)
season_2024_barreled_df['barreled_percent']  = (season_2024_barreled_df[6]/season_2024_barreled_df['lsa_events']).round(3)
```

```python
season_2024['launch_speed'] = season_2024['launch_speed'].astype(float)
```

```python
season_2024['hard_hit_ball'] = season_2024['launch_speed'].apply(lambda x: 1 if x>=95. else 0)
```
